```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/nidhipareddy/Desktop/uchicago/WA_Fn-UseC_-HR-Employee-Attrition (1).csv")
```

```python
print("Dataset Overview:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe(include="all"))

##looks like dont have nulls
```

```python
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```

```python
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object']).columns

#numerical plot columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#catagorical varables plots
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col], order=df[col].value_counts().index, palette="coolwarm")
    plt.xticks(rotation=45)
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# 7. Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 8. Boxplots to Identify Outliers
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python

if 'Attrition' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Attrition', data=df, palette='coolwarm')
    plt.title("Attrition Count")
    plt.show()

    # Attrition vs Other Categorical Columns
    for col in cat_cols:
        if col != "Attrition":
            plt.figure(figsize=(6, 4))
            sns.countplot(x=df[col], hue=df["Attrition"], palette="coolwarm")
            plt.xticks(rotation=45)
            plt.title(f'Attrition by {col}')
            plt.show()
```
